### <font color='blue'> Structured Query Language (SQL) </font>

* SQL stands for Structured Query Language.  
* We use this language to manage data in a database.  
* This allows to manage the data using tables. And we write queries using this language to fetch the data.  
* All the DBs (MsSQL, MySQL, Oracle, DB2, PostgreSQL, etc) support this language.   
* It means that a query written on one DB will also work on other DB(s). Barring some extensions that are specific to DB(s).  
* This has been made possible as ISO made SQL a database standard.
