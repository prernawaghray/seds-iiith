    contigious memory / chunk of memory
    size of array
    int arr[100]

Linked list
    fragmented
    build it element after element
    consumption of mem == num of elements


Building blocks:
    nodes - store the data
    link/reference to the next/previous node

head of linked list
    starting point of the list
    
optionally maintain the tail of list
    ending point of list


Operations
    Create
    add elements
    delete elements
    update element
    insert element
    search for element
    search for a location (indexing) - inefficient compared to array
    traverse/browse

    count of element
    size of linked list

    sort
    merge two linked lists
    split

SLL - next
DLL - next,prev
CSLL - circular ie tail and head are connected
CDLL - circular ie tail and head are connected

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
    def __del__(self):
        del (self.data)
```

```python
class SLL:
    def __init__(self):
        self.head = None
        self.size = 0
    
    def insertAtHead(self,data): #O(1)
        temp = Node(data)
        temp.next = self.head
        self.head = temp
        self.size += 1
    
    def traverse(self): # O(n) where n is number of elements
        curr = self.head
        while curr != None :
            print(curr.data,end=',')
            curr = curr.next
        print()
    
    def index(self,loc): #arr[loc] equivalent which is O(n) but in array it is O(1)
        if loc < 0 : raise Exception("location is -ve")
        if loc >= self.size : raise Exception("location out of bound")
        count = 0
        curr = self.head
        while count != loc and curr.next != None:
            count += 1
            curr = curr.next
        return curr
    
    def search(self,val): #O(n) - linear search
        curr = self.head
        while curr != None and curr.data != val: curr = curr.next
        if curr==None : raise Exception("list is empty or value not found")
        return curr

    def insertAfterLoc(self,loc,data): #O(n)
        curr = self.index(loc)
        temp = Node(data)
        temp.next = curr.next #copy the reference of what curr points to
        curr.next = temp #update the reference to temp in curr
        self.size += 1
    
    def insertBeforeLoc(self,loc,data): #O(n)
        curr = self.index(loc-1)
        temp = Node(data)
        temp.next = curr.next #copy the reference of what curr points to
        curr.next = temp #update the reference to temp in curr
        self.size += 1
    
    def deleteAtHead(self): #O(1)
        if self.head : 
            temp = self.head
            self.head = self.head.next
            del(temp) #del the Node, the destructor will take care
        else : raise Exception("list is empty")
        self.size -= 1

    def deleteAfterLoc(self,loc): #O(n)
        curr = self.index(loc)
        temp = curr.next
        curr.next = curr.next.next
        del(temp)
        self.size -= 1
    
    def deleteAfterValue(self,val): #O(n)
        curr = self.search(val)
        temp = curr.next
        curr.next = curr.next.next
        del(temp)
        self.size -= 1
    
    def merge(self,sll):
        curr = self.head
        while curr.next != None : curr = curr.next
        curr.next = sll.head


```

```python
class DLLNode:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
    def __del__(self):
        del (self.data)
```

```python
class DLL:
    def __init__(self):
        self.head = None
        self.size = 0
        self.tail = None
    
    def insertAtHead(self,data):
        temp = DLLNode(data)
        temp.next = self.head
        if self.head : self.head.prev = temp
        if self.tail==None : self.tail = temp #update once when list is empty and new node is added
        self.head = temp
        self.size += 1
    
    def traverse(self):
        curr = self.head
        tail = self.tail
        while curr != None :
            print(curr.data,end=',')
            curr = curr.next
        print()
        # print('Traversing backwards')
        # while tail != None :
        #     print(tail.data,end=',')
        #     tail = tail.prev
        # print()
    
    def search(self,val):
        curr = self.head
        while curr != None and curr.data != val : curr = curr.next
        if curr==None : raise Exception("list is null or val not found")
        return curr
    
    def insertAfterValue(self,val,data):
        curr = self.search(val)
        temp = DLLNode(data)
        temp.prev = curr
        temp.next = curr.next
        if curr != self.tail: curr.next.prev = temp
        else : self.tail = temp
        curr.next = temp
        self.size += 1
    
    def delAtHead(self):
        if self.head == None : raise Exception("empty list")
        temp = self.head.next
        if self.size > 1 : self.head.next.prev = None
        self.head = self.head.next
        del(temp)
        self.size -= 1
    
    def delBeforeValue(self,val):
        curr = self.search(val)
        temp = curr.prev
        #check for temp = head 
        temp.prev.next = curr
        curr.prev = temp.prev
        del(temp)


```
