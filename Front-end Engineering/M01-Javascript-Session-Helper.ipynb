The aim for working on this problem statement is to use multiple HTML tags as well as CSS properties for a single use case. By performing this task, you will be able to create your own portfolio website. 

The idea is to give you some starter code which you can refer. We will also create it from scratch in the session and add one part of the app at a time to make you understand.

Apart from having just HTML and CSS files now we will also have a portfolio.js file which will contain all the javascript code and we will import this as a script in our portfolio.html file after the body.

Now u must be wondering the CSS import was at the start of the html in the `<head>` tag, then why we need to have javascript imported at the end ? The reason behind it is hidden in our javascript code.
When we define a global element which is not inside any function it gets executed when we import our javascript file.

Take for example the code below:
`var monthly_salary_form = document.getElementById("form_monthly_salary");`

Here the variable `monthly_salary_form` is getting a handle of the form element which has id `form_monthly_salary`.
If we add our javascript import in the `<head>` tag then this line will fail because the html code for this form sits inside the body tag which is not yet executed. Hence we import javascript after defining the body of html.


### 1. Monthly In hand Salary Calculator

This calculator takes 3 inputs from the user. Yearly CTC, the deductions done under section 80C and a boolean input to check if the user stays in a Metro city or not. We need all these to calculate Basic salary, PF, Gratuity, Total Tax and In hand monthly salary.
    
    We use the below formula to calculate each:
    
    We read the values of ctc, deductions and is_metro from the form.
    and then calculate below 
    
    `basic = 0.5 * ctc;
    employment_tax = 2500;
    standard_deductions = 50000;
    gratuity = 0.0481 * basic;
    pf = 0.15 * basic;`
    
    The HRA is dependent on the is_metro field, If the user stays in a Metro city then the HRA is 50% of basic else it is 40% of basic salary.
    
    The tax calculation is done using below format:
    
    <pre>Income tax -> 10% on 0-5L
              -> 20% on 5-10L
              -> 30% on taxable salary above 10L</pre>
               
   The taxable salary is calculated by :
   
   `taxable = ctc - standard_deductions - employment_tax - hra - deductions - gratuity - (2*pf);`

   Using this taxable salary and the income tax slabs mentioned above we calculate the total incometax.
   
   Finally, the in hand yearly salary is calculated by 
   
   `in_hand_yearly = ctc - employment_tax - (2*pf) - gratuity - tax;`
   
   And then the in_hand_monthly salary is calculated by dividing the in_hand_yearly by 12.
