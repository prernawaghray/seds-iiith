## <font color='blue'> Online Analytical Processing </font>

Online analytical processing (OLAP) is part of the broader category of business intelligence, which also encompasses:

* Relational databases
* Report writing, and 
* Data mining

Typical applications of OLAP include: 

* Business reporting for sales & marketing
* Management reporting
* Business process management
* Bbudgeting and forecasting
* Financial reporting

OLAP tools enable users to analyze multidimensional data interactively from multiple perspectives:

OLAP consists of three basic analytical operations:

* Consolidation: Accumulating and computing data in one or more dimensions
* Drill-down: Navigating through details
* Slicing and dicing: Extract a specific set of data, and examine it from different viewpoints

### <font color='blue'> OLAP Cube </font>

An OLAP cube is a multi-dimensional array of two- or three-dimensional spreadsheet.

For example, a company might wish to summarize financial data by its dimensions: 

* Product
* Time-period
* City

to compare actual and budget expenses.

A Slice is a subset of the data, generated by picking a value for one dimension, and only showing the data for that value.

Each cell of the cube holds a number that represents some measure of the business, such as sales, profits, expenses, budget and forecast.

## <font color='blue'> Clickhouse OLAP Architecture </font>

ClickHouse is a fast open-source OLAP database management system.

It is column-oriented and allows to generate analytical reports using SQL queries in real-time.

It ingests all your structured data into the system and instantly makes it available for building reports.

ClickHouse supports multi-master asynchronous replication and can be deployed across multiple datacenters. 

ClickHouse DBMS can be configured as a purely distributed system located on independent nodes.
